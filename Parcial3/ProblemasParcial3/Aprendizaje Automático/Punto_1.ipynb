```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

# Plano tangente

```python
f = lambda x,y : x**2 - y**2 + 2*x

Dx = lambda f,x,y,h=1e-5: (f(x+h,y) - f(x-h,y))/(2*h)
Dy = lambda f,x,y,h=1e-5: (f(x,y+h) - f(x,y-h))/(2*h)

a = 1
b = 1
Poly_Taylor = lambda x,y: f(a,b) + Dx(f,a,b)*(x-a) + Dy(f,a,b)*(y-b)
```

```python
l = 10
x = np.linspace(-l,l,100)
y = np.linspace(-l,l,100)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
Zt = Poly_Taylor(X,Y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d',elev = 30, azim = 70)
ax.set_xlim3d(x[0],x[-1])
ax.set_ylim3d(y[0],y[-1])
ax.plot_surface(X,Y,Z, cmap = 'coolwarm',alpha=1)
ax.plot_surface(X,Y,Zt,color='black',alpha=0.5)
ax.scatter3D(a,b,f(a,b),color='r')
ax.set_xlabel('x')
ax.set_ylabel('y')
```
