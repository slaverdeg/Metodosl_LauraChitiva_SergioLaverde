Punto 7 Álgebra Lineal 

```python
A = np.array([[3, -1, -1],
              [-1, 3, 1],
              [2, 1, 4]])
b = np.array([1, 3, 7])

omegas = np.linspace(0, 2, 200)

def sor_solver(A, b, omega, tol=1e-6, max_iter=1000):
    n = len(b)
    x = np.zeros(n)
    
    for _ in range(max_iter):
        x_new = np.copy(x)
        for i in range(n):
            s1 = sum(A[i, j] * x_new[j] for j in range(i))
            s2 = sum(A[i, j] * x[j] for j in range(i+1, n))
            x_new[i] = (1 - omega) * x[i] + (omega / A[i, i]) * (b[i] - s1 - s2)
        if np.linalg.norm(b - np.dot(A, x_new)) < tol:
            return x_new
        x = x_new
    return x

def find_optimal_omega(A, b, tol=1e-6, max_iter=1000):
    omegas = np.linspace(0, 2, 200)  
    num_iterations = []

    for omega in omegas:
        x = sor_solver(A, b, omega, tol, max_iter)
        num_iterations.append(len(x))

    optimal_omega = omegas[num_iterations.index(min(num_iterations))]
    min_iterations = min(num_iterations)

    return optimal_omega, min_iterations

soluciones = sor_solver(A,b, omega)
find_optimal_omega(A,b)

optimal_omega, min_iterations = find_optimal_omega(A, b)
print("Las soluciones del sistema de ecuaciones es:", soluciones)
print("Número mínimo de iteraciones:", min_iterations)
```

Punto 2 Aprendizaje automatizado 

El nuevo código tiene el término de momentum adicionalmente en cada iteración. Este término se le suma al paso de descenso de gradiente.

```python
def f2(x,y):
    return x**4 + y**4 - 2*(x-y)**2

Dx = lambda f,x,y,h=1e-5: (f(x+h,y) - f(x-h,y))/(2*h)
Dy = lambda f,x,y,h=1e-5: (f(x,y+h) - f(x,y-h))/(2*h)

x0, y0 = 0.5,0.1

Gradient = lambda f,x,y: np.array([Dx(f,x,y),Dy(f,x,y)])
Gradient(f2,x0,y0)

def Minimizer(f, N = 100, gamma = 0.001, eta = 0.7):
    r = np.zeros((N,2))
    r[0] = np.array([0.0, 3.0])
    
    Grad = np.zeros((N,2))
    Grad[0] = Gradient(f,r[0,0],r[0,1])

    momentum = np.zeros((N, 2))

    for i in tqdm(range(1,N)):
        grad = Gradient(f, r[i - 1, 0], r[i - 1, 1])
        momentum[i] = eta * (r[i - 1] - r[i - 2])
        r[i] = r[i-1] - gamma*Gradient(f,r[i-1,0],r[i-1,1])
        Grad[i] = Gradient(f,r[i-1,0],r[i-1,1])
        Grad[i] = grad
    return r,Grad

Minimizer(f2)
```
